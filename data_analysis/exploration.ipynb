```python
import pandas as pd

# Load the CSV into a pandas DataFrame
df = pd.read_csv('pisos.csv')

# Display the first few rows of the DataFrame
df.head()
```

```python
# Check the data types and non-null counts
df.info()
# Setting the float format for display
pd.set_option('display.float_format', '{:.2f}'.format) 

# Basic statistical summary for numerical columns
df.describe()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print(categorical_cols)
print(len(categorical_cols))

to_filter_categorical = ["esquina", "luz", "se_aceptan_mascotas", "no_se_aceptan_mascotas"]
categorical_clean = [col for col in categorical_cols if len(df[col].value_counts()) < 15 and col not in to_filter_categorical]
print(categorical_clean)
for col in categorical_clean:
    print(df[col].value_counts())
    print('-' * 50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in categorical_clean:
    sns.countplot(y=df[col])
    plt.title(f"Frequency of {col}")
    plt.show()
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
numerical_cols_clean = [col for col in numerical_cols if col not in ["id", "createdat", "1"]]
print(numerical_cols_clean)
```

```python
for col in numerical_cols_clean:
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
correlation_matrix = df[numerical_cols_clean].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
